```python
result = []
#for tweet in tweepy.Cursor(api.search, q = "narendramodi", lang = "en", since = "2018-10-1" ).items(200):
tweet = api.user_timeline(screen_name= "@narendramodi", count = 200)  #maximum 200 are allowed 
    #print(tweet.created_at, tweet.text)
for t in tweet:
    result.append(t.text)
    
#api.GetUserTimeline(Screen_name= "")
result
```

```python
#filtering links
result2 = []
for i in range(0, len(result)):
    txt = ' '.join(word for word in result[i].split() if not word.startswith("https:"))  #removes link
    result2.append(txt)
result3 = ' '.join(word for word in result2)
result3
```

```python
#import modules for word cloud
from wordcloud import WordCloud, STOPWORDS
from PIL import Image
import numpy as np
```

```python
def create_word_cloud(text):
    
    mask = np.array(Image.open("modi.PNG"))
    stopwords = set(STOPWORDS)
    wc = WordCloud(background_color = 'white', mask= mask, max_words = 200, stopwords = stopwords)
    wc.generate(text)
    wc.to_file(os.path.join('C:\\Users\\anand','namo-words.png'))
```

```python
#calling the above funtion for wordcloud
create_word_cloud(result3)
```
